```python
#   Codigo que implementa un esquema numerico
#   para determinar la precision de una maquina
#
#           Autor:
#   kimberly mranda rosel
#   kimberlymiranda080805@mal.com
#   Version 2.0 : 03/02/2025
#

import numpy as np
import matplotlib.pyplot as plt
#Esta parte del código prepara un entorno para realizar cálculos y gráficos de una manera eficientemente ya que el numpy es una librería que utilizamos
#para poder realizar cálculos numéricos mientras que el matplotlib.pyplot es una librería que usamos para crear graficas en 2D




epsilon = 1.0         #En estas dos primeras líneas estamos dándole diciéndole al código que la iteración empieza de 0 hasta el número que no se pueda hacer más cálculos y épsilon tiene el valor de 1.0 ya que el número se seguirá dividiendo entre 2 hasta que ya no pueda dividirse mas
iteracion = 0
epsilons = []         #Las otras dos líneas son unas listas que están vacías las cuales estamos usando para poder guardarlos valores de épsilon y las iteraciones
iteraciones = []



while 1.0 + epsilon != 1.0:  # El while es el bucle que estamos utilizando con la condición que siga siendo verdadera siente y cuando 1.0 + épsilon no sea igual a 1.0
    epsilon /= 2       #La épsilon/=2 lo estamos usando ya que cada iteración se reduce a la mitad este pequeño código hace que disminuya el valor
    iteracion = iteracion + 1 #La iteración= iteración +1 es un contador que se incrementa de 1 en 1
    iteraciones.append(iteracion)   #Las iteraciones.append se agregó para ver el valor actual de la lista iteraciones es útil para tener un seguimiento de cuantas iteraciones fueron
    epsilons.append(epsilon)    #La epsilons.append se agregó para tener un seguimiento de cuánto va bajando el valor de épsilon




    print(f"Iteracion: {iteracion}, Precisión de máquina: {epsilon}")  #esta parte imprime todos las iteraciones de la 1 al 53



epsilon *= 2
print(f"Precisión de máquina: {epsilon}") #aqui estamos imprimiendo la precicion final de la maquina





plt.figure()  # este dice que sera una imagen
plt.plot(iteraciones, epsilons, label='epsilon', marker='o') #esto lo use parael demostrar por donde va la arista
plt.xscale('linear')  # se usa linear para una escala lineal
plt.yscale('linear')
plt.xlabel('epsilon') #variale x
plt.ylabel('iteraciones') #variable y
plt.legend()  #este muestra la grafica
plt.title('precision de la maquina') # este le da titulo a la grafica
plt.show()
```
