```python
# @title Texto de ttulo predeterminado
#   Codigo que implementa un esquema numerico
#   para determinar la aproximacion de Leibniz
#
#           Autor:
#   Kimberly Miranda Rosel
#   kimberlymiranda080805@gmail.com
#   Version 2.0 : 04/02/2025
#

import numpy as np
import matplotlib.pyplot as plt
#Esta parte del cdigo prepara un entorno para realizar clculos y grficos de una manera eficientemente ya que el numpy es una librera que utilizamos
#para poder realizar clculos numricos mientras que el matplotlib.pyplot es una librera que usamos para crear graficas en 2D



def leibniz_pi(n):   # esta funcion sirve para calcular una aproximcion de pi
    return 4 * sum((-1)**k / (2*k + 1) for k in range(n))




true_pi = np.pi  # esta parte almacena el verdadero valor de pi
N_values = [10, 100, 1000, 10000] # esta crea una lista con diferentes valores de n esos valores representan cuantos terminos de la serie se utilizaran para sacar una aproximacion de pi
errors_abs = [] # estas dos lineas son listas vacias que estamos utilizando para alacenar las variables
errors_rel = []
errors_cua = []




for N in N_values:
    approx_pi = leibniz_pi(N)  # Aproximacin de pi con N trminos
    error_abs = abs(true_pi - approx_pi) # Error absoluto: diferencia entre el valor verdadero de pi y la aproximacin
    error_rel = error_abs / true_pi # Error relativo: el error absoluto dividido por el valor verdadero de pi
    error_cua = error_abs**2 # Error cuadrtico: el error absoluto elevado al cuadrado
    errors_abs.append(error_abs) # Almacenamos los errores en las listas correspondientes
    errors_rel.append(error_rel)
    errors_cua.append(error_cua)
    print(f"N={N}: Error absoluto={error_abs}, Error relativo={error_rel}, Error cuadratico={error_cua}") # Imprimimos los errores para cada valor de N



plt.figure() # este dice que sera una imagen
plt.plot(N_values, errors_abs, label='Error absoluto', marker='o')#esto lo use parael demostrar por donde va la primera arista
plt.plot(N_values, errors_rel, label='Error relativo', marker='s')#esto lo use parael demostrar por donde va la segunda arista
plt.plot(N_values, errors_cua, label='Error cuadratico', marker='^')#esto lo use parael demostrar por donde va la tercera arista
plt.xscale('log')  #se usa log cuando se quiere una escala logaritmica
plt.yscale('log')
plt.xlabel('N')# variable x nombre
plt.ylabel('Error') #variable y nombre
plt.legend() #este muestra la grafica
plt.title('Errores en la aproximacin de pi') #aqui se le pone nombre a la maquina
plt.show()
```
