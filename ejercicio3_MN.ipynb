```python
#   Codigo que implementa el calculo de errores
#   en operaciones numericas
#
#           Autor:
#   Kimberly miranda rosel
#   kimberlymiranda080805@gmail.com
#   Version 1.0 : 04/02/2025
#



import numpy as np
import matplotlib.pyplot as plt
#Esta parte del código prepara un entorno para realizar cálculos y gráficos de una manera eficientemente ya que el numpy es una librería que utilizamos
#para poder realizar cálculos numéricos mientras que el matplotlib.pyplot es una librería que usamos para crear graficas en 2D



def calcular_errores(x, y, valor_real): # Definimos la función para calcular los errores
    diferencia = x - y # Calculamos la diferencia entre los valores aproximados x y y
    error_abs = abs(valor_real - diferencia) # Calculamos el error relativo, que es el error absoluto dividido por el valor real
    error_rel = error_abs / abs(valor_real)# Calculamos el error porcentual, multiplicando el error relativo por 100
    error_pct = error_rel * 100


     # Imprimimos los valores calculados para la diferencia y los errores
    print(f"Diferencia: {diferencia}")
    print(f"Error absoluto: {error_abs}")
    print(f"Error relativo: {error_rel}")
    print(f"Error porcentual: {error_pct}%")
     # Devolvemos los tres tipos de error para su uso posterior
    return error_abs, error_rel, error_pct



# Datos de ejemplo: lista de tuplas con valores (x, y, valor_real)
valores = [(1.0000001, 1.0000000, 0.0000001),
           (1.000000000000001, 1.000000000000000, 0.000000000000001)]

# Listas vacias para almacenar los errores
errores_abs = []
errores_rel = []
errores_pct = []



for x, y, real in valores: # Iteramos sobre los datos de ejemplo para calcular los errores
    print(f"\nPara x={x}, y={y}:")# Imprimimos los valores de x y y que se están analizando
    # Llamamos a la función para calcular los errores
    error_abs, error_rel, error_pct = calcular_errores(x, y, real)# Añadimos los errores calculados a las listas correspondientes
    errores_abs.append(error_abs)
    errores_rel.append(error_rel)
    errores_pct.append(error_pct)



fig, ax = plt.subplots(figsize=(10, 6)) # Creación del gráfico para visualizar los errores
indices = np.arange(len(valores)) # Creamos un arreglo de índices para usar como etiquetas en el eje X del gráfico
ax.plot(indices, errores_abs, label="Error Absoluto", marker='o', color='b') # Graficamos los errores con diferentes colores y marcadores
ax.plot(indices, errores_rel, label="Error Relativo", marker='o', color='g')
ax.set_title('Errores Absoluto, Relativo y Porcentual') # Configuración de título y etiquetas de los ejes
ax.set_xlabel('Índice de los valores')
ax.set_ylabel('Valor de los errores')
ax.set_xticks(indices) # Definimos las posiciones de las marcas en el eje X
ax.set_xticklabels([f'x={x}, y={y}' for x, y, real in valores]) # Usamos los valores de x y y para etiquetar las marcas en el eje X
ax.legend() # Añadimos la leyenda para identificar las líneas en el gráfico

plt.tight_layout() # Ajustamos el diseño del gráfico para evitar solapamientos
plt.show() # Mostramos el gráfico generado
```
